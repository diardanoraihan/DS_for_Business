# Model Building
The model type that we will build in this task is clustering model to group similar customers and assign them to a cluster. This is useful when the data is unlabeled and the process is unsupervised.

## Kmeans 
Kmeans algorithm is a type of partitioning clustering that divides the data into 'K' non-overlapping subnest (clusters) without any cluster-internal structure or labels.

Determining the number of clusters are hard, and oftenly ambiguous since it is highly dependent on the shape and scale of the distribution points in a datasets.

One of the main commonly technique to choose is to run the clustering accross the different values of K and looking at a metric of accuracy for clustering. The metric is named Within Cluster Sum of Squares (WCSS). Looking at the change of this metric, we can find the best value for K. This technique is called __The Elbow Method__.

### The Elbow Method

```python
from sklearn.cluster import KMeans
```

```python
score_i = []
k_values = range(1,20)
for i in k_values:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(cc_df_scaled)
    # retrieve the score of WCSS / within cluster sum of square  
    score_i.append(kmeans.inertia_)
```

```python
# plot the wcss vs number of cluster
# show blue dot 'bo' and connect it with a dash '-'
plt.plot(k_values, score_i, 'bo-')
plt.title('Finding the right number of clusters')
plt.xlabel('Clusters')
plt.ylabel('Score WCSS')
plt.show()
```

- From here, we can observe that cluster 4 seems to form the elbow of the curve
- However, the values doesn't reduce linearly until it reachs the 8th cluster.
- Let's choose the number of clusters to be 7.

### Kmeans Model

```python
kmeans = KMeans( n_clusters=8)
kmeans.fit(cc_df_scaled)
labels_1 = kmeans.labels_
```

```python
print(labels_1)
print(labels_1.shape)
print(labels_1.min())
print(labels_1.max())
```

```python
kmeans.cluster_centers_
```

### Insert the Labels to the Dataframe

```python
cc_df_cluster  = pd.concat([cc_df, pd.DataFrame({'cluster': labels_1})], axis=1)
cc_df_cluster
```

```python
columns = cc_df.columns

# Plot the histogram on various cluster
for col in columns:
    # create a horizontal figure (width = 35, height = 5)
    plt.figure( figsize=(35,5))
    for group in range(8):
        # in the figure, it will consist of 8 plots horizontally alligned
        plt.subplot(1, 8, group+1)
        cluster = cc_df[cc_df_cluster.cluster == group]
        cluster[col].hist( bins=20)
        plt.title('{} \nCluster {}'.format(col, group))
plt.show()
```

Using this plot, now you are able to kind of understand the different clusters, their spending habits, and their features, and characteristics as well.

## Principle Component Analysis
- PCA is unsupervised Machine Learning algorithm that performs dimensionality reductions while at attempting at keeping the original information unchanged. 
- PCA works by trying to find a new set of features called components.
- Components are composites of the uncorrelated input features. 

Let's take a look at what these clusters look like when you reduce the features.

### PCA Model

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA( n_components=2)
pca_model = pca.fit_transform( cc_df_scaled)
pca_model
```

```python
# create a dataframe with two components
pca_df = pd.DataFrame( data=pca_model, columns=['pca_1', 'pca_2'])
pca_df.head()
```

### Insert the Labels from KMeans to the PCA DF

```python
pca_df = pd.concat( [pca_df, pd.DataFrame({'cluster': labels_1})], axis=1)
```

```python
pca_df
```

### Visualize the PCA with Scatter Plot

```python
plt.figure( figsize=(10,10))
sns.scatterplot(x=pca_df.pca_1, 
                y=pca_df.pca_2, 
                hue=pca_df.cluster, 
                palette=['red', 'green','blue', 'pink', 'yellow', 'grey', 'purple', 'black'])
plt.show()
```

Basically, what we do is that we combine PCA with the cluster results from Kmeans to do our visualization.

## Auto Encoders
- Auto encoders are type of neural networks that are used to perform a task of data encoding (representation learning).
- Auto encoders use the same input data for the input and output.
<img src='Images/Autoencoder_structure.png' width='500px'>
- Using this network, we can simply perform some sort of dimensionality reduction or encoding. 
- Think of it as a pipe, and you're creating a bottleneck in the middle. Then, you're trying to reconstruct the exact input data in the output as well.

Hence, when we train this to encoders, we actually use the exact same input as the output as well. It works in the unsupervised fashion, pretty much the same idea with the principle component analysis. 


The difference is that in auto encoder, it tries to reduce (compress) its dimensions and keep the same information but represented in a much smaller space. Then, we scale it up again until we reconstruct our input again.

### The Bottleneck (Code Layer)

- Auto encoder works by adding bottleneck in the network.
- The bottleneck forces the network to create a compressed (encoded) version of the original input.
- Auto encoder works well if correlation exists between input data (performs poorly if all the input data is independent / no correlation).
- Auto encoder works great when the input is pretty much similar to the output. __X ~ X'__

__Tied Weight__: the weights from the input to the hidden layer (code layer) will be equal to the weights from the hidden layer to the output. 

### Auto Encoder Model

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Model
```

```python
input_model = Input( shape=(17,))
x = Dense(7, activation='relu')(input_model)
x = Dense(500, activation='relu', kernel_initializer='glorot_uniform')(x)
x = Dense(500, activation='relu', kernel_initializer='glorot_uniform')(x)
x = Dense(2000, activation='relu', kernel_initializer='glorot_uniform')(x)

encoded = Dense(10, activation='relu', kernel_initializer='glorot_uniform')(x)

x = Dense(2000, activation='relu', kernel_initializer='glorot_uniform')(encoded)
x = Dense(500, activation='relu', kernel_initializer='glorot_uniform')(x)

decoded = Dense(17, activation='relu', kernel_initializer='glorot_uniform')(x)

# Auto encoder
autoencoder = Model(input_model, decoded)

# Encoder
encoder = Model(input_model, encoded)

autoencoder.summary()
```

As we can see, the important point of the network is that the input and output layer have the exact same dimensions. In the process, we are creating a bottleneck in the network. So now, we can take a look at the bottleneck, which is the output of the encoder network, to take a look at the actual reduced features.

```python
# Compile the 
autoencoder.compile( optimizer='adam', loss='mean_squared_error')
```

```python
autoencoder.fit(x= cc_df_scaled, y=cc_df_scaled, batch_size=128, epochs=50, verbose=1)
```

### Encoder Model
Previously, we have defined the encoder model that takes the input with the shape of 17 features to the reduced version (encoded) with 10 features. 

From there, we can take only the encoder version and apply the predict method to it. 

```python
pred = encoder.predict(cc_df_scaled)
pred
```

```python
pred.shape
```

- As you can see, we basically just encoded the input from 17 to 10 features with the same number of data. 

- What we could do right now is to apply the KMeans algorithm to this encoded features. Before we do that, how many number of clusters do we need? This is the right time to use the __Elbow method__ again!

### Use KMeans with the Predicton from Encoder

```python
score_k = []
k_values = range(1,20)
for k in k_values:
    temp = KMeans(n_clusters=k)
    temp.fit(pred)
    score_k.append(temp.inertia_)
score_k
```

```python
# plt.figure( figsize=(10,10))
plt.plot(k_values, score_k, 'bo-')
plt.title('Finding the Right Number of Clusters')
plt.xlabel('Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
plt.plot(k_values, score_k, 'bo-', label='KMeans with Encoded Features')
plt.plot(k_values, score_i, 'ro-', label='KMeans')
plt.legend()
plt.title('Finding the Right Number of Clusters')
plt.xlabel('Clusters')
plt.ylabel('WCSS')
plt.show()
```

If you remember, for the first KMeans model (red line), we picked a large number of clusters, which is 8. But in the new model (blue line), it went high and then saturated (almost linear) after cluster 4. Hence, maybe now we can pick a smaller number of clusters, like 4 or 5. 

```python
kmeans_2 = KMeans(n_clusters=4)
kmeans_2.fit(pred)
labels_2 = kmeans_2.labels_
```

```python
cc_df_cluster_2 = pd.concat([ cc_df, pd.DataFrame({'cluster': labels_2})], axis=1)
cc_df_cluster_2
```

As you can see, we have been able to even start with 17 features, use auto encoders to drop them down to 10 features, and then use principle component analysis to drop them down to two with four clusters only.
